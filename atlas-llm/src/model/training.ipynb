```python
import torch
from cs336_basics.model.optimizer import AdamW
```

```python
weights = torch.nn.Parameter(5 * torch.randn((10, 10)))
opt = AdamW([weights], lr=1e2)
for t in range(100):
    opt.zero_grad()
    loss = (weights**2).mean()
    print(loss.cpu().item())
    if loss.cpu().item() == 0: break
    loss.backward()
    opt.step()
```
